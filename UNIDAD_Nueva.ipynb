### 2. Introducción a Pandas: El Manejo de Datos en Python

Pandas es una librería fundamental en Python para el análisis y manipulación de datos. Proporciona estructuras de datos flexibles y eficientes, siendo la más importante el **DataFrame**.

```python
import random
import pandas as pd
from datetime import datetime, timedelta

def generar_datos_biologicos(num_observaciones=200, id_estudiante=None, fecha_inicio=datetime(2023, 1, 1), fecha_fin=datetime(2023, 12, 31)):
    """
    Genera un DataFrame con datos biológicos sintéticos para un curso de Python.

    Args:
        num_observaciones (int): Número de filas de datos a generar.
        id_estudiante (int, optional): ID único para sembrar el generador de números aleatorios.
                                      Si es None, no se usa una semilla fija.
        fecha_inicio (datetime): Fecha de inicio para generar observaciones.
        fecha_fin (datetime): Fecha de fin para generar observaciones.

    Returns:
        pd.DataFrame: DataFrame con los datos biológicos generados.
    """
    if id_estudiante is not None:
        random.seed(id_estudiante)

    sitios = ["bosque", "pradera", "montaña", "rio"]

    # Rangos de peso y tamaño para cada especie (media y desviación estándar aproximada)
    rangos_especies = {
        "lobo": {"peso_media": 45, "peso_std": 15, "tamaño_media": 130, "tamaño_std": 15},
        "oso": {"peso_media": 340, "peso_std": 150, "tamaño_media": 215, "tamaño_std": 30},
        "águila": {"peso_media": 5, "peso_std": 1, "tamaño_media": 190, "tamaño_std": 20},
        "zorro": {"peso_media": 8, "peso_std": 3, "tamaño_media": 75, "tamaño_std": 10},
        "ciervo": {"peso_media": 180, "peso_std": 60, "tamaño_media": 200, "tamaño_std": 25}
    }

    # Probabilidad de encontrar especies por sitio (ajustable)
    probabilidades_sitio = {
        "bosque": {"lobo": 0.3, "oso": 0.2, "zorro": 0.3, "ciervo": 0.2, "águila": 0.1}, # Suma puede ser > 1, se normaliza
        "pradera": {"ciervo": 0.4, "zorro": 0.3, "lobo": 0.2, "águila": 0.1, "oso": 0.05},
        "montaña": {"águila": 0.4, "lobo": 0.3, "zorro": 0.2, "ciervo": 0.1, "oso": 0.05},
        "rio": {"oso": 0.3, "lobo": 0.2, "ciervo": 0.2, "águila": 0.1, "zorro": 0.1}
    }

    especies_observadas = []

    for _ in range(num_observaciones):
        sitio = random.choice(sitios)

        # Seleccionar especie basado en las probabilidades del sitio
        especies_posibles = list(probabilidades_sitio[sitio].keys())
        pesos_probabilidad = list(probabilidades_sitio[sitio].values())
        observacion = random.choices(especies_posibles, weights=pesos_probabilidad, k=1)[0]


        sexo = random.choice(["Macho", "Hembra"])
        edad = random.choice(["Juvenil", "Adulto"])

        # Generar peso y tamaño usando distribución normal (clip para evitar valores negativos)
        peso = max(0, round(random.gauss(rangos_especies[observacion]["peso_media"], rangos_especies[observacion]["peso_std"]), 2))
        tamaño = max(0, round(random.gauss(rangos_especies[observacion]["tamaño_media"], rangos_especies[observacion]["tamaño_std"]), 2))

        # Generar fecha y hora aleatoria dentro del rango
        tiempo_entre_fechas = fecha_fin - fecha_inicio
        segundos_totales = tiempo_entre_fechas.total_seconds()
        segundos_aleatorios = random.randint(0, int(segundos_totales))
        fecha_hora_observacion = fecha_inicio + timedelta(seconds=segundos_aleatorios)

        # Generar temperatura aleatoria
        temperatura = round(random.uniform(10, 30), 2)

        especies_observadas.append([fecha_hora_observacion, sitio, observacion, sexo, edad, peso, tamaño, temperatura])

    df = pd.DataFrame(especies_observadas, columns=['Fecha_Hora', 'Sitio', 'Especie', 'Sexo', 'Edad', 'Peso', 'Tamaño', 'Temperatura'])

    return df

# Ejemplo de uso:
df_datos = generar_datos_biologicos(num_observaciones=300, id_estudiante=123)
print(df_datos.head())
```

### Profundizando en la Selección de Filas: `.loc` vs `.iloc`

Anteriormente vimos cómo seleccionar filas usando `.iloc` por su posición numérica.
Ahora exploraremos con más detalle `.loc` y `.iloc`.

**¿Por qué un nuevo ejemplo?** Para entender mejor la diferencia entre estos métodos,
usaremos un DataFrame más pequeño con **índices personalizados** (nombres en lugar de
números). Nuestro DataFrame de observaciones biológicas usa números como índice (0, 1, 2...),
pero muchas veces querrás usar etiquetas más descriptivas como identificadores.

Consideremos el siguiente DataFrame de ejemplo con nombres para cada fila:

```python
# Creamos un DataFrame de ejemplo con un índice personalizado
datos = {
    'Especie': ['Roble', 'Pino', 'Abeto', 'Encina'],
    'Altura_m': [20, 35, 30, 15],
    'Region': ['Norte', 'Sur', 'Este', 'Oeste']
}
df_plantas = pd.DataFrame(datos, index=['planta1', 'planta2', 'planta3', 'planta4'])

print("DataFrame de plantas con índice personalizado:")
print(df_plantas)
```

#### `.loc` (Selección por Etiqueta/Índice)

`.loc` se utiliza principalmente para seleccionar datos basándose en **etiquetas** del índice (nombres de filas) y **nombres de columnas**.

La sintaxis básica es `df.loc[etiquetas_filas, nombres_columnas]`. Puedes usar:

*   Una sola etiqueta.
*   Una lista de etiquetas.
*   Un slice de etiquetas (¡importante: el slice final está **incluido** con `.loc`!).
*   Una máscara booleana (como ya vimos para filtrar).

```python
# Seleccionar una fila por su etiqueta de índice
fila_planta2_loc = df_plantas.loc['planta2']
print("\nFila 'planta2' usando .loc:")
print(fila_planta2_loc)

# Seleccionar múltiples filas por sus etiquetas
filas_lista_loc = df_plantas.loc[['planta1', 'planta4']]
print("\nFilas 'planta1' y 'planta4' usando .loc:")
print(filas_lista_loc)

# Seleccionar un rango de filas por sus etiquetas (el final del slice SÍ está incluido)
rango_etiquetas_loc = df_plantas.loc['planta1':'planta3']
print("\nRango de filas de 'planta1' a 'planta3' (inclusive) usando .loc:")
print(rango_etiquetas_loc)

# Seleccionar filas y columnas específicas por etiquetas
datos_planta1_loc = df_plantas.loc['planta1', 'Altura_m']
print("\nAltura de 'planta1' usando .loc:")
print(datos_planta1_loc)

# Seleccionar múltiples filas y columnas por etiquetas
datos_varias_loc = df_plantas.loc[['planta1', 'planta3'], ['Especie', 'Region']]
print("\nEspecie y Región de 'planta1' y 'planta3' usando .loc:")
print(datos_varias_loc)

# Usar .loc con una máscara booleana (esto ya lo vimos, .loc es implícito)
plantas_altas_loc = df_plantas.loc[df_plantas['Altura_m'] > 25]
print("\nPlantas con Altura_m > 25 usando .loc (máscara booleana):")
print(plantas_altas_loc)
```

#### `.iloc` (Selección por Posición Entera)

`.iloc` se utiliza para seleccionar datos basándose en la **posición entera** (basada en 0) de las filas y columnas. No importa cuáles sean las etiquetas del índice o los nombres de las columnas.

La sintaxis básica es `df.iloc[posiciones_filas, posiciones_columnas]`. Puedes usar:

*   Una sola posición entera.
*   Una lista de posiciones enteras.
*   Un slice de posiciones enteras (¡importante: el slice final está **excluido** con `.iloc`, como en las listas de Python!).

```python
# Seleccionar una fila por su posición entera (la segunda fila, índice 1)
fila_posicion1_iloc = df_plantas.iloc[1]
print("\nFila en la posición 1 (la segunda fila) usando .iloc:")
print(fila_posicion1_iloc)

# Seleccionar múltiples filas por sus posiciones enteras
filas_lista_iloc = df_plantas.iloc[[0, 3]] # La primera y la cuarta fila
print("\nFilas en las posiciones 0 y 3 usando .iloc:")
print(filas_lista_iloc)

# Seleccionar un rango de filas por sus posiciones enteras (el final del slice NO está incluido)
rango_posiciones_iloc = df_plantas.iloc[0:2] # Filas en posición 0 y 1 (las dos primeras)
print("\nRango de filas de la posición 0 a 2 (exclusiva) usando .iloc:")
print(rango_posiciones_iloc)

# Seleccionar filas y columnas específicas por posiciones enteras
dato_posicion_iloc = df_plantas.iloc[0, 1] # El valor en la primera fila, segunda columna
print("\nDato en la posición [0, 1] (primera fila, segunda columna) usando .iloc:")
print(dato_posicion_iloc)

# Seleccionar múltiples filas y columnas por posiciones enteras
datos_varios_iloc = df_plantas.iloc[[0, 2], [0, 2]] # Filas 0 y 2, columnas 0 y 2
print("\nDatos en las posiciones de filas [0, 2] y columnas [0, 2] usando .iloc:")
print(datos_varios_iloc)

# Puedes usar slicing para filas y una lista para columnas, o viceversa
mezcla_iloc = df_plantas.iloc[1:3, [0, 2]] # Filas en posición 1 y 2, columnas 0 y 2
print("\nDatos mezclando slicing y lista de posiciones con .iloc:")
print(mezcla_iloc)
```

#### Resumen Clave:

*   **`.loc`**: Usa **etiquetas** (nombres de filas y columnas). Incluye el final del slice cuando se usan etiquetas de índice.
*   **`.iloc`**: Usa **posiciones enteras** (basadas en 0). Excluye el final del slice, como en el slicing de listas estándar de Python.

Comprender la diferencia entre estos dos métodos es fundamental para seleccionar datos de manera efectiva en Pandas, especialmente cuando trabajas con DataFrames que tienen índices personalizados.

### Ampliando el Filtrado de Datos: Máscaras Booleanas y `.query()`

El filtrado es una operación esencial para seleccionar subconjuntos de datos que cumplen ciertas condiciones. Ya hemos visto la forma más común de hacerlo usando **máscaras booleanas**. Ahora profundizaremos en ellas y exploraremos el método `.query()` que puede simplificar el código en algunos casos.

#### Filtrado con Máscaras Booleanas (Repaso y Más Ejemplos)

Una máscara booleana es una Serie o un array de valores `True` o `False`, donde la longitud es igual al número de filas del DataFrame. Cuando aplicamos esta máscara al DataFrame, Pandas devuelve solo las filas donde la máscara es `True`.

Las condiciones se crean utilizando operadores de comparación (`==`, `!=`, `>`, `<`, `>=`, `<=`) y operadores lógicos (`&` para AND, `|` para OR, `~` para NOT). Es crucial encerrar cada condición individual entre paréntesis `()`.

```python
# Usaremos el DataFrame original de datos biológicos
print(df_datos.head())

# Filtrar observaciones de especies que NO son 'lobo'
no_lobos = df_datos[df_datos['Especie'] != 'lobo']
print("\nObservaciones que no son lobos:")
print(no_lobos.head())

# Filtrar observaciones con Peso entre 5 y 10 kg (ambos inclusive)
peso_entre_5_y_10 = df_datos[(df_datos['Peso'] >= 5) & (df_datos['Peso'] <= 10)]
print("\nObservaciones con Peso entre 5 y 10 kg:")
print(peso_entre_5_y_10.head())

# Filtrar observaciones que son 'Oso' O que están en el sitio 'montaña'
osos_o_en_montana = df_datos[(df_datos['Especie'] == 'oso') | (df_datos['Sitio'] == 'montaña')]
print("\nObservaciones que son Oso o están en el sitio montaña:")
print(osos_o_en_montana.head())

# Filtrar observaciones de Hembras Adultas
hembras_adultas = df_datos[(df_datos['Sexo'] == 'Hembra') & (df_datos['Edad'] == 'Adulto')]
print("\nObservaciones de Hembras Adultas:")
print(hembras_adultas.head())

# Filtrar observaciones donde la Temperatura es menor a 15 O el Tamaño es mayor a 200
frio_o_grande = df_datos[(df_datos['Temperatura'] < 15) | (df_datos['Tamaño'] > 200)]
print("\nObservaciones con Temperatura < 15 o Tamaño > 200:")
print(frio_o_grande.head())
```

#### Filtrado con el Método `.query()`

El método `.query()` permite filtrar un DataFrame usando una sintaxis basada en cadenas de texto. Para filtros simples o moderadamente complejos, puede hacer que el código sea más legible.

La sintaxis es `df.query('condicion')`, donde `condicion` es una cadena que representa la expresión de filtrado. Los nombres de las columnas se pueden usar directamente dentro de la cadena. Para comparar con valores de texto, usa comillas simples dentro de la cadena de consulta (o comillas dobles si la columna tiene espacios o caracteres especiales).

```python
# Filtrar observaciones donde la Especie es 'lobo' usando .query()
lobos_query = df_datos.query("Especie == 'lobo'")
print("\nObservaciones de lobos (usando .query()):")
print(lobos_query.head())

# Filtrar observaciones donde el Peso es mayor a 200 kg Y el Sitio es 'rio' usando .query()
osos_grandes_en_rio_query = df_datos.query("Peso > 200 and Sitio == 'rio'")
print("\nObservaciones de osos grandes en río (usando .query()):")
print(osos_grandes_en_rio_query) # Mostrar todas, ya que el ejemplo anterior mostró pocas

# Filtrar observaciones con Peso entre 5 y 10 kg usando .query()
peso_entre_5_y_10_query = df_datos.query("Peso >= 5 and Peso <= 10")
print("\nObservaciones con Peso entre 5 y 10 kg (usando .query()):")
print(peso_entre_5_y_10_query.head())

# Filtrar observaciones que son 'Oso' O que están en el sitio 'montaña' usando .query()
osos_o_en_montana_query = df_datos.query("Especie == 'oso' or Sitio == 'montaña'")
print("\nObservaciones que son Oso o están en el sitio montaña (usando .query()):")
print(osos_o_en_montana_query.head())

# Filtrar observaciones de Hembras Adultas usando .query()
hembras_adultas_query = df_datos.query("Sexo == 'Hembra' and Edad == 'Adulto'")
print("\nObservaciones de Hembras Adultas (usando .query()):")
print(hembras_adultas_query.head())
```

#### Cuándo usar Máscaras Booleanas vs `.query()`

*   **Máscaras Booleanas:** Son la forma más flexible y fundamental. Funcionan bien para cualquier tipo de condición y son necesarias si la condición involucra funciones o lógica más compleja que `.query()` no puede manejar directamente. A veces, para condiciones muy simples, pueden ser un poco más verbosas.
*   **`.query()`:** Puede ser más legible para condiciones de filtrado simples a moderadas, especialmente si tienes múltiples condiciones combinadas con `and` u `or`. Es particularmente útil cuando los nombres de las columnas no contienen espacios o caracteres especiales (aunque `.query()` puede manejarlos si se encierran entre backticks `` ` ``).

Ambos métodos logran el mismo resultado: seleccionar filas basadas en condiciones. La elección entre uno u otro a menudo depende de la preferencia personal y la complejidad del filtro.

#### Agrupación y Agregación

A menudo queremos calcular estadísticas descriptivas para subgrupos de nuestros datos (por ejemplo, el peso promedio por especie o la temperatura máxima por sitio). Pandas facilita esto con el método `.groupby()`.

El proceso de "group by" generalmente implica tres pasos:

1.  **Splitting (División):** Dividir los datos en grupos basándose en algún criterio (por ejemplo, los valores únicos en una columna categórica como 'Especie' o 'Sitio').
2.  **Applying (Aplicación):** Aplicar una función a cada grupo de forma independiente. Esta función puede ser una función de agregación (como `mean()`, `median()`, `sum()`, `count()`, `min()`, `max()`, `std()`, `var()`), una transformación (que devuelve un resultado del mismo tamaño que el grupo), o un filtro.
3.  **Combining (Combinación):** Combinar los resultados de cada grupo en una estructura de datos única (generalmente un DataFrame o Serie).

El método `.groupby()` realiza los pasos de división y preparación para la aplicación. Luego, aplicas una función de agregación (o `agg()`) para realizar los pasos de aplicación y combinación.

```python
# Usaremos el DataFrame original de datos biológicos
print(df_datos.head())

# Ejemplo básico: Calcular el peso promedio por especie
#
# PASO 1 - SPLITTING (División):
#   Pandas agrupa todas las filas que tienen la misma especie
#   - Grupo "lobo": todas las filas donde Especie == 'lobo'
#   - Grupo "oso": todas las filas donde Especie == 'oso'
#   - etc.
#
# PASO 2 - APPLYING (Aplicación):
#   A cada grupo, se le aplica la función .mean() sobre la columna 'Peso'
#   - Peso promedio del grupo "lobo"
#   - Peso promedio del grupo "oso"
#   - etc.
#
# PASO 3 - COMBINING (Combinación):
#   Pandas junta todos los resultados en una Serie con el resultado por especie

peso_promedio_por_especie = df_datos.groupby('Especie')['Peso'].mean()
print("\nPeso promedio por especie:")
print(peso_promedio_por_especie)
```

Puedes agrupar por múltiples columnas pasando una lista de nombres de columnas a `groupby()`. Esto creará un índice jerárquico (MultiIndex) en el resultado.

```python
# Ejemplo: Calcular el tamaño mediano por sitio y sexo
tamaño_mediano_por_sitio_sexo = df_datos.groupby(['Sitio', 'Sexo'])['Tamaño'].median()
print("\nTamaño mediano por sitio y sexo:")
print(tamaño_mediano_por_sitio_sexo)
```

El método `.agg()` (o `.aggregate()`) te permite aplicar una o varias funciones de agregación a una o varias columnas después de agrupar. Puedes pasar una lista de nombres de funciones (como cadenas) o funciones reales.

```python
# Ejemplo: Calcular múltiples estadísticas para la columna 'Peso' por 'Edad'
# Queremos el mínimo, máximo y promedio del peso para cada grupo de edad.
estadisticas_peso_por_edad = df_datos.groupby('Edad')['Peso'].agg(['min', 'max', 'mean'])
print("\nEstadísticas de peso por edad:")
print(estadisticas_peso_por_edad)

# Puedes renombrar las columnas resultantes si lo deseas
estadisticas_peso_por_edad_renombrado = df_datos.groupby('Edad')['Peso'].agg(
    peso_minimo='min',
    peso_maximo='max',
    peso_promedio='mean'
)
print("\nEstadísticas de peso por edad (columnas renombradas):")
print(estadisticas_peso_por_edad_renombrado)
```

También puedes aplicar diferentes funciones de agregación a diferentes columnas en la misma operación `groupby().agg()`. Para esto, pasas un diccionario a `agg()` donde las claves son los nombres de las columnas a agregar y los valores son las funciones (o lista de funciones) a aplicar a esa columna.

```python
# Ejemplo: Calcular el peso promedio y la temperatura máxima por sitio
estadisticas_varias_por_sitio = df_datos.groupby('Sitio').agg(
    peso_promedio=('Peso', 'mean'), # Tupla: (columna, función)
    temperatura_maxima=('Temperatura', 'max'),
    conteo_observaciones=('Especie', 'count') # Contar observaciones por especie (o cualquier otra columna no nula)
)
print("\nEstadísticas varias por sitio:")
print(estadisticas_varias_por_sitio)

# Ejemplo: Aplicar múltiples funciones a una columna y una función a otra
estadisticas_complejas = df_datos.groupby('Especie').agg(
    peso_stats=('Peso', ['min', 'max', 'mean', 'std']), # Múltiples funciones para Peso
    tamaño_promedio=('Tamaño', 'mean') # Una función para Tamaño
)
print("\nEstadísticas complejas por especie:")
print(estadisticas_complejas)
```

Los métodos `groupby()` y `agg()` son fundamentales para realizar análisis de datos
agregados y resumir información por categorías. Con estas herramientas puedes responder
preguntas como "¿Cuál es la altura promedio de los árboles por región?" o "¿Cuántas
observaciones de cada especie se registraron en cada sitio?".

## Limpieza de Datos

Antes de realizar cualquier análisis, es crucial asegurarse de que nuestros datos estén limpios y en un formato adecuado. ### Manejo de Datos Faltantes (Missing Data)

Los valores faltantes son datos que no están disponibles o no se registraron para una observación particular. En Pandas, los valores numéricos faltantes suelen representarse como `NaN` (Not a Number), mientras que para otros tipos de datos pueden ser `None` o valores especiales como `NaT` (Not a Time) para fechas.

#### Identificar y Contar Valores Faltantes

Podemos usar el método `.isnull()` (o `.isna()`, son sinónimos) para obtener un DataFrame booleano que indica `True` donde hay un valor faltante y `False` en caso contrario.

Para contar los valores faltantes por columna, podemos encadenar `.isnull()` con `.sum()`.

#### Estrategias para Manejar Valores Faltantes

Una vez identificados los valores faltantes, tenemos varias opciones para manejarlos, dependiendo del contexto y la cantidad de datos faltantes.

1.  **Eliminar Filas o Columnas con Faltantes (`.dropna()`)**:
    *   Podemos eliminar filas que contengan al menos un valor faltante (`axis=0` o por defecto).
    *   Podemos eliminar columnas que contengan al menos un valor faltante (`axis=1`).
    *   Podemos especificar el umbral de valores no faltantes requeridos (`thresh`).
    *   Podemos especificar si eliminar si "any" (cualquiera) o "all" (todos) los valores son faltantes en la fila/columna (`how`).

```python
# Crear una copia del DataFrame para practicar la eliminación sin modificar el original
df_cleaned_drop = df_datos.copy()

# Ejemplo: Eliminar filas con al menos un valor faltante
print("Forma del DataFrame antes de eliminar filas con faltantes:", df_cleaned_drop.shape)
df_cleaned_drop_rows = df_cleaned_drop.dropna(how='any')
print("Forma del DataFrame después de eliminar filas con al menos un faltante:", df_cleaned_drop_rows.shape)
# print(df_cleaned_drop_rows.head()) # Puede que no muestre nada si se eliminan las primeras filas

# Ejemplo: Eliminar columnas con al menos un valor faltante
df_cleaned_drop_cols = df_cleaned_drop.dropna(axis=1, how='any')
print("\nForma del DataFrame antes de eliminar columnas con faltantes:", df_cleaned_drop.shape)
print("Forma del DataFrame después de eliminar columnas con al menos un faltante:", df_cleaned_drop_cols.shape)
print(df_cleaned_drop_cols.head()) # Mostrar el resultado
```

2.  **Rellenar Valores Faltantes (`.fillna()`)**:
    *   Podemos reemplazar los valores faltantes con un valor específico (ej. 0, un string, etc.).
    *   Podemos rellenar con la media, mediana o moda de la columna.
    *   Podemos usar métodos de propagación: `ffill` (forward fill, usar el último valor válido anterior) o `bfill` (backward fill, usar el próximo valor válido posterior).

```python
# Crear una copia del DataFrame para practicar el rellenado
df_cleaned_fill = df_datos.copy()

# Ejemplo: Rellenar faltantes en la columna 'Peso' con la media de esa columna
# Primero, para demostración, introducimos algunos NaNs artificialmente
import numpy as np
df_cleaned_fill.loc[df_cleaned_fill.sample(frac=0.05).index, 'Peso'] = np.nan
print("Conteo de faltantes en 'Peso' antes de rellenar:", df_cleaned_fill['Peso'].isnull().sum())

media_peso = df_cleaned_fill['Peso'].mean()
df_cleaned_fill['Peso_rellenado'] = df_cleaned_fill['Peso'].fillna(media_peso) # Crear nueva columna para comparación

print("\nConteo de faltantes en 'Peso_rellenado' después de rellenar con la media:", df_cleaned_fill['Peso_rellenado'].isnull().sum())
print(df_cleaned_fill[['Peso', 'Peso_rellenado']].head(10)) # Comparar columnas

# Ejemplo: Rellenar faltantes en una columna categórica con un valor específico (ej. 'Desconocido')
# Asumiendo que 'Sitio' pudiera tener faltantes
# df_cleaned_fill['Sitio'] = df_cleaned_fill['Sitio'].fillna('Desconocido')

# Ejemplo: Rellenar usando forward fill (útil en series de tiempo)
# df_cleaned_fill['Temperatura_ffill'] = df_cleaned_fill['Temperatura'].fillna(method='ffill')
```

**Media vs. Mediana para Rellenar Datos Faltantes:**

Cuando decides rellenar (`fillna()`) valores faltantes en una columna numérica, la elección entre usar la **media** o la **mediana** depende principalmente de la distribución de tus datos:

*   **Usar la Media (`.mean()`):**
    *   Es una buena opción cuando tus datos tienen una distribución aproximadamente **simétrica** y **no contienen valores atípicos (outliers)** significativos.
    *   La media utiliza todos los valores en el cálculo, por lo que es sensible a los extremos. Rellenar con la media puede distorsionar la distribución si hay valores extremos presentes.

*   **Usar la Mediana (`.median()`):**
    *   Es una mejor opción cuando tus datos tienen una distribución **sesgada** (asimétrica hacia un lado) o **contienen valores atípicos**.
    *   La mediana es el valor central y no se ve afectada por los valores extremos. Rellenar con la mediana mantiene mejor la forma de la distribución original en presencia de outliers.


```python
# Simulemos pesos con un valor atípico (outlier)
pesos_ejemplo = [45, 47, 46, 48, 45, 44, 46, 500]  # 500 kg es un error de medición

media_ejemplo = np.mean(pesos_ejemplo)
mediana_ejemplo = np.median(pesos_ejemplo)

print(f"Pesos: {pesos_ejemplo}")
print(f"Media: {media_ejemplo:.2f} kg")      # ~108 kg (¡distorsionado por el 500!)
print(f"Mediana: {mediana_ejemplo:.2f} kg")  # 46.5 kg (valor más representativo)

print("\n💡 Observa cómo la media se ve muy afectada por el valor atípico de 500 kg,")
print("   mientras que la mediana se mantiene cerca de los valores típicos.")
print("\n   Para datos biológicos con posibles errores de medición, la mediana")
print("   suele ser más robusta para rellenar valores faltantes.")
```


**En resumen:**

Si tus datos son limpios y simétricos, la media puede ser adecuada. Si sospechas o sabes que tus datos tienen sesgo o valores atípicos, la mediana suele ser una elección más robusta para rellenar faltantes, ya que es menos sensible a estos extremos y preserva mejor las características de la distribución central.

```python
# Crear una copia del DataFrame para practicar el rellenado con diferentes métodos
df_cleaned_fill_methods = df_datos.copy()

# Para demostración, introducimos algunos NaNs artificialmente en 'Peso' y 'Tamaño'
import numpy as np
df_cleaned_fill_methods.loc[df_cleaned_fill_methods.sample(frac=0.05).index, 'Peso'] = np.nan
df_cleaned_fill_methods.loc[df_cleaned_fill_methods.sample(frac=0.05).index, 'Tamaño'] = np.nan

print("Conteo de faltantes antes de rellenar:")
print(df_cleaned_fill_methods[['Peso', 'Tamaño']].isnull().sum())

# Ejemplo: Rellenar faltantes en la columna 'Peso' con la media de esa columna
media_peso = df_cleaned_fill_methods['Peso'].mean()
df_cleaned_fill_methods['Peso_Media_Rellenado'] = df_cleaned_fill_methods['Peso'].fillna(media_peso)

print("\nConteo de faltantes en 'Peso_Media_Rellenado' después de rellenar con la media:")
print(df_cleaned_fill_methods['Peso_Media_Rellenado'].isnull().sum())

# Ejemplo: Rellenar faltantes en la columna 'Tamaño' con la mediana de esa columna
mediana_tamaño = df_cleaned_fill_methods['Tamaño'].median()
df_cleaned_fill_methods['Tamaño_Mediana_Rellenado'] = df_cleaned_fill_methods['Tamaño'].fillna(mediana_tamaño)

print("\nConteo de faltantes en 'Tamaño_Mediana_Rellenado' después de rellenar con la mediana:")
print(df_cleaned_fill_methods['Tamaño_Mediana_Rellenado'].isnull().sum())

print("\nComparación de columnas originales y rellenadas:")
print(df_cleaned_fill_methods[['Peso', 'Peso_Media_Rellenado', 'Tamaño', 'Tamaño_Mediana_Rellenado']].head(10))
```

#### Identificar y Eliminar Duplicados

Podemos usar el método `.duplicated()` para obtener una Serie booleana que indica `True` para las filas que son duplicados (manteniendo la primera ocurrencia por defecto).

Podemos usar el método `.drop_duplicates()` para eliminar las filas duplicadas. Por defecto, mantiene la primera ocurrencia.
```python
import random
import pandas as pd
from datetime import datetime, timedelta
import numpy as np # Importar numpy para usar np.nan

def generar_datos_sucios_biologicos(num_observaciones=200, id_estudiante=None, fecha_inicio=datetime(2023, 1, 1), fecha_fin=datetime(2023, 12, 31), porcentaje_faltantes=0.08, porcentaje_duplicados=0.05):
    """
    Genera un DataFrame con datos biológicos sintéticos, INTENCIONALMENTE incluyendo
    valores faltantes y duplicados para prácticas de limpieza de datos.

    Args:
        num_observaciones (int): Número de filas de datos a generar inicialmente.
        id_estudiante (int, optional): ID único para sembrar el generador de números aleatorios.
                                      Si es None, no se usa una semilla fija.
        fecha_inicio (datetime): Fecha de inicio para generar observaciones.
        fecha_fin (datetime): Fecha de fin para generar observaciones.
        porcentaje_faltantes (float): Porcentaje aproximado de valores que serán NaN.
        porcentaje_duplicados (float): Porcentaje aproximado de filas que serán duplicadas.


    Returns:
        pd.DataFrame: DataFrame con los datos biológicos generados (incluyendo "suciedad").
    """
    if id_estudiante is not None:
        random.seed(id_estudiante)
        np.random.seed(id_estudiante) # También sembrar numpy para NaNs y duplicados

    sitios = ["bosque", "pradera", "montaña", "rio"]

    # Rangos de peso y tamaño para cada especie (media y desviación estándar aproximada)
    rangos_especies = {
        "lobo": {"peso_media": 45, "peso_std": 15, "tamaño_media": 130, "tamaño_std": 15},
        "oso": {"peso_media": 340, "peso_std": 150, "tamaño_media": 215, "tamaño_std": 30},
        "águila": {"peso_media": 5, "peso_std": 1, "tamaño_media": 190, "tamaño_std": 20},
        "zorro": {"peso_media": 8, "peso_std": 3, "tamaño_media": 75, "tamaño_std": 10},
        "ciervo": {"peso_media": 180, "peso_std": 60, "tamaño_media": 200, "tamaño_std": 25}
    }

    # Probabilidad de encontrar especies por sitio (ajustable)
    probabilidades_sitio = {
        "bosque": {"lobo": 0.3, "oso": 0.2, "zorro": 0.3, "ciervo": 0.2, "águila": 0.1}, # Suma puede ser > 1, se normaliza
        "pradera": {"ciervo": 0.4, "zorro": 0.3, "lobo": 0.2, "águila": 0.1, "oso": 0.05},
        "montaña": {"águila": 0.4, "lobo": 0.3, "zorro": 0.2, "ciervo": 0.1, "oso": 0.05},
        "rio": {"oso": 0.3, "lobo": 0.2, "ciervo": 0.2, "águila": 0.1, "zorro": 0.1}
    }

    especies_observadas = []

    for _ in range(num_observaciones):
        sitio = random.choice(sitios)

        # Seleccionar especie basado en las probabilidades del sitio
        especies_posibles = list(probabilidades_sitio[sitio].keys())
        pesos_probabilidad = list(probabilidades_sitio[sitio].values())
        observacion = random.choices(especies_posibles, weights=pesos_probabilidad, k=1)[0]


        sexo = random.choice(["Macho", "Hembra"])
        edad = random.choice(["Juvenil", "Adulto"])

        # Generar peso y tamaño usando distribución normal (clip para evitar valores negativos)
        peso = max(0, round(random.gauss(rangos_especies[observacion]["peso_media"], rangos_especies[observacion]["peso_std"]), 2))
        tamaño = max(0, round(random.gauss(rangos_especies[observacion]["tamaño_media"], rangos_especies[observacion]["tamaño_std"]), 2))

        # Generar fecha y hora aleatoria dentro del rango
        tiempo_entre_fechas = fecha_fin - fecha_inicio
        segundos_totales = tiempo_entre_fechas.total_seconds()
        segundos_aleatorios = random.randint(0, int(segundos_totales))
        fecha_hora_observacion = fecha_inicio + timedelta(seconds=segundos_aleatorios)

        # Generar temperatura aleatoria
        temperatura = round(random.uniform(10, 30), 2)

        especies_observadas.append([fecha_hora_observacion, sitio, observacion, sexo, edad, peso, tamaño, temperatura])

    df = pd.DataFrame(especies_observadas, columns=['Fecha_Hora', 'Sitio', 'Especie', 'Sexo', 'Edad', 'Peso', 'Tamaño', 'Temperatura'])

    # --- Introducir datos sucios ---

    # 1. Introducir valores faltantes (NaN) en columnas seleccionadas
    columnas_con_faltantes = ['Peso', 'Tamaño', 'Temperatura', 'Edad', 'Sitio'] # Columnas donde introducir NaNs
    for col in columnas_con_faltantes:
        if col in df.columns:
            num_faltantes_col = int(num_observaciones * porcentaje_faltantes)
            indices_a_modificar = np.random.choice(df.index, num_faltantes_col, replace=False)
            df.loc[indices_a_modificar, col] = np.nan

    # 2. Introducir filas duplicadas
    num_filas_a_duplicar = int(num_observaciones * porcentaje_duplicados)
    if num_filas_a_duplicar > 0:
        # Seleccionar filas aleatoriamente para duplicar
        filas_a_duplicar_indices = np.random.choice(df.index, num_filas_a_duplicar, replace=True)
        filas_duplicadas = df.loc[filas_a_duplicar_indices].reset_index(drop=True)
        df = pd.concat([df, filas_duplicadas], ignore_index=True)

    # Mezclar las filas para que los duplicados y NaNs no estén solo al final
    df = df.sample(frac=1, random_state=id_estudiante).reset_index(drop=True)


    return df

# Ejemplo de uso para generar datos sucios:
df_datos_sucios = generar_datos_sucios_biologicos(num_observaciones=300, id_estudiante=456, porcentaje_faltantes=0.07, porcentaje_duplicados=0.04)
print("Ejemplo de DataFrame con datos sucios:")
print(df_datos_sucios.head())

print("\nVerificación de datos 'sucios':")
print("Conteo de valores faltantes por columna:")
print(df_datos_sucios.isnull().sum())
print("\nNúmero de filas duplicadas (excluyendo la primera ocurrencia):")
print(df_datos_sucios.duplicated().sum())
```

### 3. Estadística Descriptiva Básica con Pandas

Una vez que tenemos nuestros datos cargados en un DataFrame de Pandas, podemos comenzar a realizar análisis descriptivos para entender las características principales de nuestras observaciones.

La estadística descriptiva nos ayuda a resumir y visualizar los datos para obtener información clave. Pandas proporciona funciones muy útiles para esto.

#### Medidas de Tendencia Central

Las medidas de tendencia central nos indican el valor típico o central de un conjunto de datos. Las más comunes son la media, la mediana y la moda.

*   **Media (`.mean()`):** Es el promedio de los valores.
*   **Mediana (`.median()`):** Es el valor central cuando los datos están ordenados. Es menos sensible a valores atípicos que la media.
*   **Moda (`.mode()`):** Es el valor que aparece con mayor frecuencia. Un conjunto de datos puede tener una moda, varias modas o ninguna.

```python
# Calcular la media del peso
media_peso = df_datos['Peso'].mean()
print(f"La media del peso observado es: {media_peso:.2f} kg")

# Calcular la mediana del tamaño
mediana_tamaño = df_datos['Tamaño'].median()
print(f"La mediana del tamaño observado es: {mediana_tamaño:.2f} cm")

# Calcular la moda de las especies
moda_especie = df_datos['Especie'].mode()
print(f"La especie más observada es:\n{moda_especie}")

# Calcular la moda del sitio
moda_sitio = df_datos['Sitio'].mode()
print(f"El sitio con más observaciones es:\n{moda_sitio}")
```

#### Medidas de Dispersión

Las medidas de dispersión nos indican qué tan dispersos o variados están los datos alrededor de la tendencia central.

*   **Varianza (`.var()`):** Mide la dispersión promedio de los valores con respecto a la media. Su unidad está al cuadrado de la unidad original.
*   **Desviación Estándar (`.std()`):** Es la raíz cuadrada de la varianza e indica la dispersión promedio en la misma unidad que los datos originales. Es muy utilizada.
*   **Rango:** La diferencia entre el valor máximo y mínimo. Aunque no hay una función `.range()` directa en Pandas, se puede calcular fácilmente.

```python
# Calcular la desviación estándar del peso
desviacion_estandar_peso = df_datos['Peso'].std()
print(f"La desviación estándar del peso es: {desviacion_estandar_peso:.2f} kg")

# Calcular la varianza del tamaño
varianza_tamaño = df_datos['Tamaño'].var()
print(f"La varianza del tamaño es: {varianza_tamaño:.2f} cm²")

# Calcular el rango de la temperatura
rango_temperatura = df_datos['Temperatura'].max() - df_datos['Temperatura'].min()
print(f"El rango de la temperatura es: {rango_temperatura:.2f} °C")
```

#### Resumen Estadístico Completo

Pandas ofrece la función `.describe()` que proporciona un resumen estadístico muy útil para columnas numéricas, incluyendo conteo, media, desviación estándar, mínimo, cuartiles (25%, 50%, 75%) y máximo.

```python
# Obtener el resumen estadístico de las columnas numéricas
resumen_numerico = df_datos[['Peso', 'Tamaño', 'Temperatura']].describe()
print(resumen_numerico)
```

Para columnas no numéricas (categóricas), `.describe()` proporciona el conteo, el número de valores únicos, el valor más frecuente (`top`) y su frecuencia (`freq`).

```python
# Obtener el resumen estadístico de las columnas categóricas
resumen_categorico = df_datos[['Sitio', 'Especie', 'Sexo', 'Edad']].describe()
print(resumen_categorico)
```

#### Tablas de Frecuencia

Las tablas de frecuencia nos muestran cuántas veces aparece cada valor único en una columna.

*   **`.value_counts()`:** Calcula la frecuencia de cada valor único en una Serie.
*   **`pd.crosstab()`:** Permite crear tablas de contingencia para mostrar la frecuencia conjunta de dos o más variables categóricas.

```python
# Tabla de frecuencia de especies
conteo_especies = df_datos['Especie'].value_counts()
print("Conteo de especies observadas:")
print(conteo_especies)

# Tabla de frecuencia de sitios
conteo_sitios = df_datos['Sitio'].value_counts()
print("\nConteo de observaciones por sitio:")
print(conteo_sitios)

# Tabla de contingencia de Sitio vs Especie
tabla_contingencia_sitio_especie = pd.crosstab(df_datos['Sitio'], df_datos['Especie'])
print("\nTabla de contingencia de Sitio vs Especie:")
print(tabla_contingencia_sitio_especie)

# Tabla de contingencia de Sexo vs Edad
tabla_contingencia_sexo_edad = pd.crosstab(df_datos['Sexo'], df_datos['Edad'])
print("\nTabla de contingencia de Sexo vs Edad:")
print(tabla_contingencia_sexo_edad)
```

#### Agrupación y Agregación

A menudo queremos calcular estadísticas descriptivas para subgrupos de nuestros datos (por ejemplo, el peso promedio por especie o la temperatura máxima por sitio). Pandas facilita esto con el método `.groupby()`.

```python
# Peso promedio por especie
peso_promedio_por_especie = df_datos.groupby('Especie')['Peso'].mean()
print("Peso promedio por especie:")
print(peso_promedio_por_especie)

# Tamaño mediano por sitio y sexo
tamaño_mediano_por_sitio_sexo = df_datos.groupby(['Sitio', 'Sexo'])['Tamaño'].median()
print("\nTamaño mediano por sitio y sexo:")
print(tamaño_mediano_por_sitio_sexo)

# Múltiples estadísticas por grupo (Peso mínimo, máximo y promedio por Edad)
estadisticas_peso_por_edad = df_datos.groupby('Edad')['Peso'].agg(['min', 'max', 'mean'])
print("\nEstadísticas de peso por edad:")
print(estadisticas_peso_por_edad)
```

Con estos ejemplos básicos puedes empezar a explorar y resumir tus datos utilizando
las herramientas de Pandas.

Una vez que hemos resumido nuestros datos con estadística descriptiva, el siguiente paso crucial es visualizarlos. La visualización nos permite identificar patrones, tendencias, valores atípicos y relaciones entre variables de una manera intuitiva.

Usaremos dos librerías principales en Python para visualización:

*   **Matplotlib:** Es una librería fundamental para crear gráficos estáticos de alta calidad.
*   **Seaborn:** Es una librería construida sobre Matplotlib que proporciona una interfaz de alto nivel para crear gráficos estadísticos atractivos e informativos. A menudo simplifica la creación de gráficos complejos.

```python
# Importar las librerías de visualización
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración básica para mejorar la apariencia de los gráficos
sns.set_style("whitegrid") # Estilo de cuadrícula blanca
plt.rcParams['figure.figsize'] = (10, 6) # Tamaño por defecto de las figuras
```

### 5. Introducción al Análisis de Series de Tiempo con Pandas

Los datos biológicos a menudo tienen un componente temporal. El análisis de series de tiempo nos permite examinar patrones, tendencias y estacionalidades en los datos a lo largo del tiempo. Pandas, combinado con Matplotlib y Seaborn, ofrece herramientas poderosas para esto.

#### Asegurando el Formato de Fecha y Hora

Es crucial que la columna de fecha y hora esté en el formato correcto (`datetime`) para que Pandas pueda realizar operaciones de tiempo eficientes. Nuestra función `generar_datos_biologicos` ya lo hace, pero si cargas datos externos, a veces necesitarás convertir una columna de texto a datetime usando `pd.to_datetime()`.

```python
# Verificar el tipo de dato de la columna 'Fecha_Hora'
print("Tipo de dato de 'Fecha_Hora' antes de asegurar el formato:")
print(df_datos['Fecha_Hora'].dtype)

# Si no fuera datetime, usaríamos:
# df_datos['Fecha_Hora'] = pd.to_datetime(df_datos['Fecha_Hora'], errors='coerce')
# El argumento errors='coerce' convertirá los valores que no pueda parsear a NaT (Not a Time).

print("\nTipo de dato de 'Fecha_Hora' después de asegurar el formato:")
print(df_datos['Fecha_Hora'].dtype)
```

#### Establecer la Columna de Fecha/Hora como Índice

A menudo es útil establecer la columna de fecha y hora como el índice del DataFrame. Esto facilita la selección de rangos de tiempo y el uso de funcionalidades de series de tiempo.

```python
# Crear una copia del DataFrame para no modificar el original si no se desea
df_time = df_datos.copy()

# Establecer 'Fecha_Hora' como el índice
df_time = df_time.set_index('Fecha_Hora')

print("\nDataFrame con 'Fecha_Hora' como índice:")
print(df_time.head())
```

#### Selección de Datos por Rango de Tiempo

Con el índice de fecha y hora, es muy sencillo seleccionar datos dentro de un período específico.

```python
# Seleccionar todas las observaciones en un mes específico (Ej: Enero de 2023)
enero_2023 = df_time['2023-01']
print("\nObservaciones en Enero de 2023:")
print(enero_2023.head())

# Seleccionar observaciones en un rango de fechas específico
rango_fechas = df_time['2023-03-15':'2023-04-30']
print("\nObservaciones entre el 15 de Marzo y el 30 de Abril de 2023:")
print(rango_fechas.head())

# Seleccionar observaciones en un día específico
dia_especifico = df_time.loc['2023-07-20'] # Usamos .loc con el índice de fecha
print("\nObservaciones en el día 20 de Julio de 2023:")
print(dia_especifico)
```

#### Re-muestreo de Datos Temporales

El re-muestreo (resampling) nos permite cambiar la frecuencia de nuestra serie temporal, por ejemplo, para agregar datos por día, semana o mes. Esto es muy útil para analizar tendencias a largo plazo.

```python
# Contar el número de observaciones por día
conteo_diario = df_time.resample('D').size()
print("\nConteo de observaciones por día:")
print(conteo_diario.head())

# Calcular el peso promedio por semana
peso_promedio_semanal = df_time['Peso'].resample('W').mean()
print("\nPeso promedio observado por semana:")
print(peso_promedio_semanal.head())

# Calcular la temperatura máxima por mes
temp_max_mensual = df_time['Temperatura'].resample('M').max()
print("\nTemperatura máxima observada por mes:")
print(temp_max_mensual.head())

# Contar la frecuencia de cada especie por mes
# Primero, necesitamos una columna dummy para contar
df_time['count'] = 1
conteo_especies_mensual = df_time.groupby('Especie').resample('M')['count'].sum().unstack(level=0)
print("\nConteo mensual de cada especie:")
print(conteo_especies_mensual.head())

# Nota: .unstack() reorganiza el DataFrame para que las especies sean columnas
# Puedes llenar los valores faltantes (meses sin observaciones) con 0
conteo_especies_mensual = conteo_especies_mensual.fillna(0)
print("\nConteo mensual de cada especie (NaN rellenados con 0):")
print(conteo_especies_mensual.head())
```

#### Visualización de Series de Tiempo

Visualizar los datos a lo largo del tiempo es clave para identificar patrones.

```python
# Gráfico de la temperatura promedio por día
conteo_diario.plot(figsize=(12, 6))
plt.title('Conteo Diario de Observaciones')
plt.xlabel('Fecha')
plt.ylabel('Número de Observaciones')
plt.show()

# Gráfico del peso promedio semanal
peso_promedio_semanal.plot(figsize=(12, 6), color='orange')
plt.title('Peso Promedio Semanal de Observaciones')
plt.xlabel('Fecha')
plt.ylabel('Peso Promedio (kg)')
plt.show()

# Gráfico del conteo mensual de cada especie
conteo_especies_mensual.plot(figsize=(14, 8))
plt.title('Conteo Mensual de Observaciones por Especie')
plt.xlabel('Fecha')
plt.ylabel('Conteo')
plt.legend(title='Especie')
plt.show()

# Otro ejemplo: Temperatura promedio por día, coloreada por sitio
# Primero, calculamos la temperatura promedio diaria por sitio
temp_promedio_diaria_sitio = df_datos.groupby('Sitio').resample('D', on='Fecha_Hora')['Temperatura'].mean().unstack(level=0)
temp_promedio_diaria_sitio = temp_promedio_diaria_sitio.fillna(method='ffill') # Rellenar NaNs con el valor anterior

print("\nTemperatura promedio diaria por sitio:")
print(temp_promedio_diaria_sitio.head())

# Visualizar
temp_promedio_diaria_sitio.plot(figsize=(14, 7))
plt.title('Temperatura Promedio Diaria por Sitio')
plt.xlabel('Fecha')
plt.ylabel('Temperatura Promedio (°C)')
plt.legend(title='Sitio')
plt.show()
```

Estos ejemplos demuestran cómo puedes manipular y visualizar datos biológicos que varían a lo largo del tiempo. El re-muestreo y la agrupación son técnicas muy útiles para analizar tendencias temporales en diferentes categorías.
